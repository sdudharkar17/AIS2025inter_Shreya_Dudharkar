```python
19
import pandas as pd

# Load your data file into a DataFrame
df = pd.read_csv('/content/Daily Weather.csv')

# Find numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Print numerical columns
print("Numerical columns:")
print(numerical_columns)

# Choose a numerical column
column_name = input("Enter the name of the numerical column: ")

# Check if the column exists
if column_name in numerical_columns:
    # Find the minimum and maximum value of the column
    min_value = df[column_name].min()
    max_value = df[column_name].max()

    print(f"Minimum value in '{column_name}': {min_value}")
    print(f"Maximum value in '{column_name}': {max_value}")
else:
    print("The column does not exist or is not numerical.")
```

```python
20
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load your data file into a DataFrame
df = pd.read_csv('/content/Daily Weather.csv')

# Find categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Print categorical columns
print("Categorical columns:")
print(categorical_columns)

# Choose a categorical column
column_name = input("Enter the name of the categorical column: ")

# Check if the column exists
if column_name in categorical_columns:
    # Create a LabelEncoder object
    le = LabelEncoder()

    # Apply label encoding to the column
    df[f"{column_name}_encoded"] = le.fit_transform(df[column_name])

    print(f"\nDataFrame after label encoding '{column_name}':")
    print(df[[column_name, f"{column_name}_encoded"]].head())

    # Print the mapping of categories to encoded values
    print("\nMapping of categories to encoded values:")
    for i, category in enumerate(le.classes_):
        print(f"{category}: {i}")
else:
    print("The column does not exist or is not categorical.")
```
