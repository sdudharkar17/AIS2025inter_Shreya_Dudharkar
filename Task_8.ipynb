```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate some sample data
np.random.seed(0)
data = np.random.normal(0, 1, 100)
data = np.append(data, [5, 6, 7])  # Add some outliers

# Box and Whisker Plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, vert=False)
plt.title("Box and Whisker Plot")
plt.show()

# Find Quartile
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
print("Quartile 1 (Q1):", Q1)
print("Quartile 3 (Q3):", Q3)
print("Interquartile Range (IQR):", IQR)

# Tukey's Method
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data < lower_bound) | (data > upper_bound)]
print("Outliers (Tukey's Method):", outliers)

# Z Score
mean = np.mean(data)
std_dev = np.std(data)
z_scores = [(x - mean) / std_dev for x in data]
outliers = [x for x, z in zip(data, z_scores) if abs(z) > 2]
print("Outliers (Z Score):", outliers)

# Standardization
standardized_data = [(x - mean) / std_dev for x in data]

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(range(len(data)), data)
plt.title("Scatter Plot")
plt.show()
```
