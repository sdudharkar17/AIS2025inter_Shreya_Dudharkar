```python
import seaborn as sns
df=sns.load_dataset("tips")
df.head()
```

```python
df.corr
```

```python
df.dtypes
```

```python
sns.jointplot(x="tip",y="total_bill",data=df)
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="hex")
sns.jointplot(x="tip",y="total_bill",data=df,kind="reg")
```

```python
sns.pairplot(df)
sns.pairplot(df,hue="sex")
sns.pairplot(df,hue="day")
sns.pairplot(df,hue="total_bill")
```

```python
sns.displot(df['tip'])
```

```python
sns.countplot(df['sex'])
sns.countplot(df['day'])
sns.countplot(df['time'])
sns.countplot(df['smoker'])
```

```python
sns.boxplot(x='smoker',y='total_bill',data=df,palette='rainbow')
```

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
